```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
df= pd.read_csv('netflix1.csv')
df.head()
```

```python
# df.shape
print(f'Data has {df.shape[1]} columns and {df.shape[0]} rows.')
```

```python
df.drop_duplicates()
```

```python
# Data Information
df.info()
```

```python
# For extracting null value
# True = null values are in the data set.
# False = No null are in the data set.
df.isnull()
```

```python
# To evaluate missing value in the form of numeric.
df.isnull().sum()
```

```python
# plt.figure(figsize= (20, 20))
# sns.heatmap(df.isnull())
```

```python
df.columns
```

```python
df.drop(columns='show_id', axis=1, inplace=True)
df.columns
```

```python
# Looking How many TV shows or Movies are in the dataset.

type= df['type'].value_counts().reset_index()
type
```

```python
df.nunique()
```

```python
dir = df.groupby(['director', 'type']).size().reset_index(name='count')
dir
```

```python

top_10_dir= dir[(dir['director'] != "Not Given")& (dir['director'] != "")].sort_values(by='count', ascending=False).head(10)
top_10_dir
```

```python
countries= df.groupby(['country', 'type']).size().reset_index(name='count')
countries
```

```python
top_10_country= countries.sort_values(by='count', ascending=False).head(10)
top_10_country
```

```python
years= df['release_year'].value_counts
years
```

```python
ratings= df.groupby(['rating', 'type']).size().reset_index(name='count')
ratings.head(10)
```

```python
top_10_ratings= ratings.sort_values(by= 'count', ascending= False).reset_index().head(10)
top_10_ratings
```

```python
duration= df.groupby(['duration', 'type']).size().reset_index(name= 'count')
duration
```

```python
top_5_dur= duration.sort_values(by= 'count', ascending= False).reset_index().head(10)
top_5_dur
```

```python
list= df.groupby(['listed_in', 'type']).size().reset_index(name='count')
list
```

```python
top_10_list= list.sort_values(by= 'count', ascending= False).reset_index().head(10)
top_10_list
```

```python
countries_list= df.groupby(['country', 'listed_in']).size().reset_index(name= 'count')
# countries_list
top_10= countries_list.sort_values(by= 'count', ascending= False).reset_index().head(10)
top_10
```

                                      Visualization

```python

plt.figure(figsize=(10, 6))
plt.pie(type['type'], labels=type['index'], autopct='%1.1f%%', startangle=90, shadow=True,
        pctdistance= 0.50)
plt.title('Distribution Of Shows Types', y=1.12, fontsize=15)
plt.axis('equal')
plt.legend(labels=type['index'], loc='upper left', fontsize=7)
plt.show()
```

```python

plt.figure(figsize=(14, 6))
plt.hist(top_10_country['country'],bins= 80,
         color='coral', alpha=0.8)
plt.xlabel('Country')
plt.ylabel('Counts')
plt.title('Top 10 Distribution of Countries')
plt.xticks(rotation=75)  # Rotate x-axis labels for better readability

plt.show()
```

```python
indian_directors = df[df['country'] == 'India']

director_counts = indian_directors['director'].value_counts().reset_index()
director_counts.columns = ['Director', 'Count']

director_counts = director_counts.sort_values(by='Count', ascending=False).head(20)


# top_10_directors = director_counts.head(10)
print(director_counts)
```

```python
plt.figure(figsize=(12, 8))
plt.barh(director_counts['Director'], director_counts['Count'], color='b')
plt.xlabel('Number of Movies/Shows')
plt.ylabel('Director')
plt.title('Top 10 Indian Directors on Netflix')
plt.gca().invert_yaxis()  # Invert the y-axis to show the highest count at the top

plt.show()
```

```python
usa_directors = df[df['country']== 'United States']

directors_cou= usa_directors['director'].value_counts().reset_index()
directors_cou.columns =['Director', 'Count']

directors_cou = directors_cou.sort_values(by='Count', ascending=False).head(20)
print(directors_cou)

plt.figure(figsize=(12, 8))
plt.barh(directors_cou['Director'], directors_cou['Count'], color='b')
plt.xlabel('Number of Movies/Shows')
plt.ylabel('Director')
plt.title('Top 10 United Directors on Netflix')
plt.gca().invert_yaxis()
```

```python

usa_movie_directors = df[(df['country'] == 'United States') & (df['type'] == 'Movie')]
directors_list = usa_movie_directors['director'].unique()
print(directors_list[:10])
```

```python
plt.figure(figsize=(20, 8))
sns.barplot(data=top_10_list,x='listed_in',y='count',hue='type')
plt.ylabel('Number of Shows')
plt.xlabel('List of shows', fontsize=16)
plt.xticks(rotation = 90)
plt.title('Top 10 list of shows')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_ratings,x='rating',y='count',hue='type')
plt.ylabel('Number of Shows')
plt.xlabel('Ratings', fontsize=16)
plt.title('Top 10 Genre')
```

```python

plt.figure(figsize= (15,6))
plt.pie(data=top_5_dur,x='count',labels='duration',autopct='%.1f%%', startangle= 90, shadow= True,
        pctdistance=0.50)
plt.legend(top_5_dur['duration'], loc= 'upper left')
plt.axis('equal')
plt.title('Top 5 Durations', y= 1.12)
plt.show()
```

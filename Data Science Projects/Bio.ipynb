```python
import pandas as pd
import numpy as np
from chembl_webresource_client.new_client import new_client
```

## Search for Target Protien
#### Target search for coronavirus

```python
target = new_client.target
target_query= target.search('coronavirus')
targets= pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target= targets.target_chembl_id[6]
selected_target
```

#### Here, we will retrive only bioactivity data for coronavirus 3C- like proteinase (CHEMBL3927)  that are reported as IC50 values in nM(nanomolar) unit.

```python
activity= new_client.activity
res= activity.filter(target_chembl_id= selected_target).filter(standard_type= 'IC50')
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head()
```

```python
df.standard_type.unique()
```

```python
# save the result of bioactivity data to a CSV file
df.to_csv('bioactivity_data.csv', index= False)
```

## Handling Missing Values

```python
df1= df[df.standard_value.notna()]
df1
```

There is no missing data into the bioactivity dataset

### Data pre-processing of the bioactivity data

```python
bioactivity_class= []
for i in df1.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append('inactive')
    elif float(i) <= 1000:
        bioactivity_class.append('active')
    else:
        bioactivity_class.append('intermeadiate')    
```

```python
df1.molecule_chembl_id
```

```python
# Iterate the molecule_chembl_id to a list
mol_cid= []
for i in df1.molecule_chembl_id:
    mol_cid.append(i)
```

```python
mol_cid
```

```python
# Iterate canonical_smiles to a list
canonical_smiles= []
for i in df1.canonical_smiles:
    canonical_smiles.append(i)
```

```python
canonical_smiles
```

```python
# Iterate standard_values to a list
standard_value= []
for i in df1.standard_value:
    standard_value.append(i)
```

```python
standard_value
```

```python
# Altenative methods
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df2= df1[selection]
df2
```

```python
pd.concat([df2, pd.Series(bioactivity_class)], axis= 1)
```

```python
# Combine the 4 lists into a dataframe
data_tuples= list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df2 = pd.DataFrame(data_tuples, columns= ['molecule_chembl_id','bioactivity_class', 'canonical_smiles', 'standard_value'])
df2
```

```python
df2.to_csv('bioactivity_preprocessed_data.csv', index= False)
```

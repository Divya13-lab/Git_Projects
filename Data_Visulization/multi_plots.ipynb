```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv('tips.csv')
df.head()
```

## Distplot
### Distplot (sns.distplot()) combines the histogram & plots the estimated probability density function over the data.
#### --- Bin size is calculated using square root of row counts.

```python
sns.displot(df['total_bill'], bins= 16, color= 'green', kde= True)
```

```python
# for change the kde color(, kde=True, kde_kws={'color': 'red'})
# displot does not directly accept kde_kws for the kde parameter, seaborn.
# histplot allows for more direct control over KDE customization.
# sns.histplot(df['total_bill'], bins=16, color='green', kde= True)
# sns.kdeplot(df['total_bill'], color='red', lw=2)
```

## Joint plot
- A joint plot takes two variables and create histograme and scatter plot together. 

```python
sns.jointplot(x= 'total_bill', y= 'tip', data= df, color= 'green', kind= 'scatter',
              joint_kws= {'color': 'blue'}, marginal_kws= {'hist_kws':{'color': 'green'}})
```

```python
# sns.jointplot(x= 'total_bill', y= 'tip', data= df, kind= 'hex', color= 'lightcoral',
            #   joint_kws= {'color': 'blue'}, marginal_kws= {'hist_kws':{'color': 'green'}})

sns.jointplot(x= 'total_bill', y= 'tip', data= df, kind= 'hex',
              joint_kws= {'color': 'blue'}, marginal_kws={'color': 'red', 'bins': 30})
```

```python
sns.jointplot(x= 'total_bill', y= 'tip', data= df, kind= 'hex',
              color= 'blue', marginal_kws={'color': 'red', 'bins': 30})
```

### Joint plot: kind= 'kde' Type

```python
# sns.jointplot(x= 'total_bill', y= 'tip', data= df, kind= 'kde', color= 'blue')
sns.jointplot(x= df['total_bill'], y= df['tip'], kind= 'kde', color= 'blue', fill= True,
              kde_kws={'shade': True, 'alpha': 0.5})
```

## Pair-Plot
- A pairplot essentially plots pair wise relationship between variables.
- Hue colors the plot using categorical columns.

```python
sns.pairplot(df, hue= 'sex', palette= 'dark')
```

## Bar plot
- For ploting categorical columns Vs numerical columns.

```python
# Let's create a bar plot of 'total_bill' with 'sex', and see that who pays more, male or female.
sns.barplot(x= 'sex', y= 'total_bill', data= df)
```

```python
sns.barplot(x= 'day', y= 'total_bill', data= df)
```

## Box plot (For categorical-continous variable)
- A box plot is a visual representation of five point summary statistic of given dataset. 
- A five number summary includes:- Minimum, First Quartile, Median(second quantile), Third quartile and Maximum.

```python
sns.boxplot(x= 'day', y= 'total_bill', data= df)
```

```python
sns.boxplot(x= 'day', y= 'total_bill', data= df, hue= 'smoker')
```

## Violin Plot
- For catgorical continuous variables
- Violin plots are similar to the Box plot.

```python
sns.violinplot(x= 'day', y= 'total_bill', data= df)
```

## LM Plot
#### LM Plot is the best fits the regression line to the dataset showing as a scatter plots.
- It follows the ordinary least square method and the line represent the best fit line.

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= df, hue= 'day')
```
